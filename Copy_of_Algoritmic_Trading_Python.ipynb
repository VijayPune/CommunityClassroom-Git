### Now we will focus our analysis on only one stock, say Apple and find the Returns on Apple stock

```python
aapl = close.AAPL.copy().to_frame()
```

```python
aapl.describe()
```

```python
aapl["lag1"] = aapl.shift(periods=1)     # To shift all data by one row down. Used to find differences between consecutive values in a column
```

```python
aapl        # To calculate % Change in Close on daily basis, we find difference between 2 consecutive date Close value using AAPL & lag1
```

```python
aapl["diff"] = aapl.AAPL.sub(aapl.lag1)    # Find the Difference and save in a new column Diff
aapl
```

```python
aapl["% Change"] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
aapl
```

```python
# Another method instead of using Substract method, can use "dif" method
aapl["Diff2"] = aapl.AAPL.diff(periods=1)
aapl
```

```python
aapl["% Change 2"] = aapl.AAPL.pct_change(periods=1).mul(100)
aapl
```

```python
del aapl["diff"]     # Remove the unwanted columns
aapl
```

```python
aapl.rename(columns = {'% Change 2' : 'Change'}, inplace = True)    # To renmae a column name pass as Dictionary (key,value) pair
aapl
```

```python
# Let us take only last date of the month data instead of daily data of Close values
aapl.AAPL.resample("M").last()    # To get Monthly sample
```

```python
# Let us take only last BUSINESS DAY of the month data instead of daily data of Close values
aapl.AAPL.resample("BM").last()    # Business Day is the last day on which trading happened, no Sunday/Holiday
```

```python
# We can also find % Change on a Monthly Basis, it will be 6.53% or -1.16 % month over month
aapl.AAPL.resample("BM").last().pct_change(periods=1).mul(100)
```
